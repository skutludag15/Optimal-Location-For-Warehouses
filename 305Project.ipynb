```python
import time

with open('test/test8') as f:
    lines = f.read().splitlines()
    counter = 0
    for line in lines:
        if(counter == 0):
            n, k = line.split(" ")
            lst = [[None,None] for _ in range(0,int(n))]
        else:
            loc, weight = line.split(" ")
            lst[counter-1] = [int(loc), int(weight)]
        counter = counter + 1

start = time.time()
lstWare = [[None,None] for _ in range(0,int(k))]
lst2 = [False for _ in range(0,int(n))]

for i in range(0,int(k)):
    minimum = 999999
    min1 = 0
    for ind1 in range(0,int(n)):
        if(lst2[ind1] == False):
            whouse1 = lst[ind1]
            ans = 0
            for m in range(0,len(lst)):
                valid = False
                if(m==ind1):
                    ans = ans + 0                    
                else:
                    for wareInd in range(0,len(lstWare)):
                        whousePast = lstWare[wareInd]
                        if(whousePast[0] != None):
                            if(abs(whousePast[0]-lst[m][0])<abs(whouse1[0]-lst[m][0])):
                                valid = True
                    if(valid):
                        ans = ans + 0
                    else:
                        ans = ans + abs(whouse1[0]-lst[m][0]) * lst[m][1]
            if(ans < minimum):
                minimum = ans
                min1 = ind1
    lstWare[i] = lst[min1] 
    lst2[min1] = True
    
ans = 0
for i in range(0,int(n)):
    road = 9999
    if(lst2[i]== False):
        for m in range(0, int(k)):
            if(abs(lst[i][0]-lstWare[m][0])<road):
                road = abs(lst[i][0]-lstWare[m][0])                
        ans = ans + road * lst[i][1]
    else:
        ans = ans + 0

finish = time.time()
print("Time"+str(finish-start))
print(ans)        
```

```python
from itertools import combinations
import time

with open('test/test9') as f:
    lines = f.read().splitlines()
    counter = 0
    for line in lines:
        if(counter == 0):
            n, k = line.split(" ")
            lst = [[None,None] for _ in range(0,int(n))]
        else:
            loc, weight = line.split(" ")
            lst[counter-1] = [int(loc), int(weight)]
        counter = counter + 1

start = time.time()
lstWare = [[None,None] for _ in range(0,int(k))]
minimum = 9999999
comb = combinations(range(0,int(n)), int(k))

for c in comb:
    ans = 0
    for i in range(0,int(k)):
        wHouse = lst[c[i]]
        lstWare[i] = wHouse
    for ind in range(0,int(n)):
        factory = lst[ind]
        road=9999999999
        for wareNum in range(0,int(k)):
            ware = lstWare[wareNum]
            if(abs(ware[0]-factory[0])*factory[1]<road):
                road = abs(ware[0]-factory[0])*factory[1]
        ans = ans + road
    if(ans < minimum):
        minimum = ans
        
finish = time.time()
print("Time"+str(finish-start))
print(minimum)                
```
