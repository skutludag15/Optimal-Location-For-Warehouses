```python
from itertools import combinations
import time

with open('test/test20') as f:
    lines = f.read().splitlines()
    counter = 0
    for line in lines:
        if(counter == 0):
            n, k = line.split(" ")
            lst = [[None,None] for _ in range(0,int(n))]
        else:
            loc, weight = line.split(" ")
            lst[counter-1] = [int(loc), int(weight)]
        counter = counter + 1

start = time.time()
lstWare = [[None,None] for _ in range(0,int(k))]
minimum = 9999999
comb = combinations(range(0,int(n)), int(k))

for c in comb:
    ans = 0
    for i in range(0,int(k)):
        wHouse = lst[c[i]]
        lstWare[i] = wHouse
    for ind in range(0,int(n)):
        factory = lst[ind]
        road=9999999999
        for wareNum in range(0,int(k)):
            ware = lstWare[wareNum]
            if(abs(ware[0]-factory[0])*factory[1]<road):
                road = abs(ware[0]-factory[0])*factory[1]
        ans = ans + road
    if(ans < minimum):
        minimum = ans
        
finish = time.time()
print("Time"+str(finish-start))
print(minimum)                
```

```python
import time

def get1Median(lst):
    answer = 0
    weight = 0
    for fact in lst:
        answer = answer + fact[0]*fact[1]
        weight = weight + fact[1]
    answer = answer / weight
    index = 0
    counter = 0
    road = 999999
    for fact in lst:
        if(abs(answer-fact[0])<road):
            road = abs(answer-fact[0])
            index = counter 
        counter = counter + 1
    return index

def get2Median(lst):
    ind = get1Median(lst)
    size1 = int(ind)
    size2 = len(lst) - size1 - 1
    lst1 = [[None,None] for _ in range(0,size1)]
    lst2 = [[None,None] for _ in range(0,size2)]
    count = 0
    count1 = 0
    for fact in lst:
        if(count < size1):
            lst1[count] = [fact[0], fact[1]]
            count = count + 1
        elif(count > size1):
            lst2[count1] = [fact[0], fact[1]]
            count1 = count1 + 1
        else:
            count = count + 1
    return lst1, lst2
        
        
with open('test/test2') as f:
    lines = f.read().splitlines()
    counter = 0
    for line in lines:
        if(counter == 0):
            n, k = line.split(" ")
            lst = [[None,None] for _ in range(0,int(n))]
        else:
            loc, weight = line.split(" ")
            lst[counter-1] = [int(loc), int(weight)]
        counter = counter + 1


start = time.time()
lstWare = [[None,None] for _ in range(0,int(k))]
wareNum = int(k)

while(True):
    if(wareNum == 1):
        lol1 = get1Median(lst)
        lstWare[len(lstWare)-wareNum] = [lst[lol1][0],lst[lol1][1]]
        wareNum = wareNum - 1 
        break
    elif(wareNum == 2):
        lst2, lst3 = get2Median(lst)
        ex1 = get1Median(lst2)
        ex2 = get1Median(lst3)
        lstWare[len(lstWare)-wareNum] = [lst2[ex1][0],lst2[ex1][1]]
        wareNum = wareNum - 1
        lstWare[len(lstWare)-wareNum] = [lst3[ex2][0],lst3[ex2][1]]
        wareNum = wareNum - 1
        break
    elif(wareNum % 2 == 0):    
        lst2, lst3 = get2Median(lst)
        ex1 = get1Median(lst2)
        ex2 = get1Median(lst3)
        lstWare[len(lstWare)-wareNum] = [lst2[ex1][0],lst2[ex1][1]]
        wareNum = wareNum - 1
        lstWare[len(lstWare)-wareNum] = [lst3[ex2][0],lst3[ex2][1]]
        wareNum = wareNum - 1
    elif(wareNum % 2 == 1):
        lol1 = get1Median(lst)
        lstWare[len(lstWare)-wareNum] = [lst[lol1][0],lst[lol1][1]]
        wareNum = wareNum - 1

ans = 0
for i in range(0,int(n)):
    road = 999999
    for m in range(0, int(k)):
        if(abs(lst[i][0]-lstWare[m][0])<road):
            road = abs(lst[i][0]-lstWare[m][0])                
    ans = ans + road * lst[i][1]

finish = time.time()
print("Time"+str(finish-start))
print(ans)        
```
